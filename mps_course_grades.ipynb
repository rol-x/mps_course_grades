```python
"""
Jupyter notebook with summary of the MPS subject grades, done in Python.

MPS (Microprocessor Systems) course was divided into 5 laboratories, with three
to four tasks in each lab. The grades are real, but the names are changes.
"""
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the grades
sheet_id = '1a2DYLUOzGZYriArN-shquYYraearR6cy67Aa688ry5c'
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv'
grades = pd.read_csv(url).drop('L.p.', axis=1)
```

```python
# Slice the names off
names = grades['Name']
grades.drop('Name', axis=1, inplace=True)
grades
```

```python
# Create hierarchical index for labs and tasks
lab_labels = []
task_labels = []
# How many tasks in each lab
lab_tasks = {1: 4, 2: 3, 3: 3, 4: 3, 5: 3}
for lab in lab_tasks:
    for task in range(lab_tasks[lab]):
        lab_labels.append(f'L{lab}')
        task_labels.append(f'T{task+1}')

ind_tuples = list(zip(*[lab_labels, task_labels]))
index = pd.MultiIndex.from_tuples(ind_tuples, names=['Lab', 'Task'])
grades.columns = index
```

```python
# Create matching index for names, merge names back with grades
names.name = ('', 'Name')
grades = pd.concat([names, grades], axis=1)
```

```python
# Drop empty values, reset the index
grades.dropna(inplace=True)
grades = grades.set_index(('', 'Name')).rename_axis(index='Name')
```

```python
grades
```

```python
# Sum task grades in each lab
performance = grades.groupby(level=0, axis=1).sum()
```

```python
total = pd.DataFrame(performance.sum(axis=1)).rename(columns={0: 'Total'})
```

```python
performance['Total'] = total
```

```python
# Calculate grades as percentage
performance['L1'] /= 0.04
performance['L2'] /= 0.03
performance['L3'] /= 0.03
performance['L4'] /= 0.03
performance['L5'] /= 0.03
performance['Total'] /= 0.16
performance[['L1', 'L2', 'L3', 'L4', 'L5', 'Total']] = performance.round(2)
```

```python
# Sort performance by total score
performance = performance.sort_values(by=['Total', 'Name'],
                                      ascending=[False, True])
performance
```

```python
# Divide students into groups (same exact scores for tasks) in given lab
def group_students(lab):
    """Gather students who scored exactly the same on given lab into groups."""
    group_grades = grades.assign(temp=grades[lab].sum(axis=1)) \
        .sort_values('temp', ascending=False)[lab]
    changes = (group_grades.pct_change()).any(axis=1)
    group_grades['Group'] = ''
    group_tag = 65
    for ind in range(0, len(group_grades)):
        name = group_grades.index[ind]
        group_grades.loc[name, 'Group'] = chr(group_tag)
        if ind < len(group_grades) - 1:
            if changes[ind + 1]:
                group_tag += 1
    # group_grades.rename_axis(columns = 'Tasks')
    return group_grades
```

```python
# Divide students into groups over all labs
def group_students_all():
    """Gather students who scored exactly the same on all labs into groups."""
    group_grades = grades.assign(temp=grades.sum(axis=1)) \
        .sort_values('temp', ascending=False).drop('temp', 1)
    changes = (group_grades.pct_change()).any(axis=1)
    group_grades['Group'] = ''
    group_tag = 65
    for ind in range(0, len(group_grades)):
        name = group_grades.index[ind]
        group_grades.loc[name, 'Group'] = chr(group_tag)
        if ind < len(group_grades) - 1:
            if changes[ind + 1]:
                group_tag += 1
            if group_tag == 90:
                group_tag = 97
    return group_grades
```

```python
# Create a dictionary of same score groups and students in them
groups = {}
all_grouped = group_students_all().groupby(['Group', 'Name']).mean()
for group_code, name in all_grouped.index:
    if group_code in groups.keys():
        groups[group_code].append(name)
    else:
        groups[group_code] = [name]
```

```python
# Display groups with more than one student (same scores on all labs)
clever_students = pd.DataFrame()
for group_code in groups:
    if len(groups[group_code]) > 1:
        clever_group = all_grouped.xs(group_code)
        clever_group['Group'] = group_code
        clever_students = pd.concat([clever_students, clever_group])

clever_students
```
